```python
def changeHospitalwithlistofChange(hospital,listofChange):
    for cord in listofChange:
        x=cord[0]
        y=cord[1]
        hospital[x][y]=2
    return hospital
        


def getAllCordinate(i,j,r,c,key):
    leftCord=[]
    rightCord=[]
    downCord=[]
    topCord=[]
    
    left=j-1
    if left>=0:
        leftCord=[i,left]
        
    right=j+1
    if right<=c-1:
        rightCord=[i,right]
        
    top=i-1
    if top>=0:
        topCord=[top,j]
        
    down=i+1
    if down<=r-1:
        downCord=[down,j]
    
    keyList=[key,0]
        
    
    return[keyList,topCord,rightCord,downCord,leftCord]

    
def helpaterp(hospital,count):
    
    r=len(hospital)
    c=len(hospital[0])
    infect=False
    listOfChange=[]
    for i in range(r):
        for j in range(c):
            allCordList=getAllCordinate(i,j,r,c,hospital[i][j])
            
            
            for idx,cord in enumerate(allCordList):
                if idx==0:
                    key=cord[0]
                else:
                    if len(cord)==2:
                        
                        x=cord[0]
                        y=cord[1]
                        print(x,y,key)
                        if key==2 and hospital[x][y]==1:
                            infect=True
                            listOfChange.append([x,y])
                            
                            
                
        
    if infect:
        count+=1
        
        hospital=changeHospitalwithlistofChange(hospital,listOfChange)
        return helpaterp(hospital,count)
    else:
        
        return count
            
```

```python
class OptimizeSolution:
    
    def calc(self,hospital):
        r=len(hospital)
        c=len(hospital[0])
        count=0
        infected=[]
        not_infected=0
        for i in range(r):
            for j in range(c):
                key=hospital[i][j]
                if key==2:
                    infected.append([i,j])
                elif key==1:
                    not_infected+=1
        
        #print("not_inf= ",not_inf,"infected len= ",infected)
        while not_infected>0 and len(infected)>0:
            #print("not_inf start= ",not_inf,"infected len= ",infected)
            count+=1
            infectedNew=[]
            for f in infected:
                i=f[0]
                j=f[1]
                #print("i",i,"j",j)
                #left side
                left=j-1
                if left>=0:
                    if hospital[i][left]==1:
                        hospital[i][left]=2
                        infectedNew.append([i,left])
                        
                #right side        
                right=j+1
                if right<=c-1:
                    if hospital[i][right]==1:
                        hospital[i][right]=2
                        infectedNew.append([i,right])
                    
        
                #top side    
                top=i-1
                if top>=0:
                    if hospital[top][j]==1:
                        hospital[top][j]=2
                        infectedNew.append([top,j])
                    
                #down side
                down=i+1
                if down<=r-1:
                    if hospital[down][j]==1:
                        hospital[down][j]=2
                        infectedNew.append([down,j])
            
            
                
            infected=infectedNew
            not_infected-=len(infectedNew)
            
        
    
        if not_infected>0:
            return -1
            
        return count
        
    def helpaterp(self, hospital):
        # code here
        return self.calc(hospital)
```
